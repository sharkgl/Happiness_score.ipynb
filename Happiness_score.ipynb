```python
grouped = df[df['Happiness Score'] >= 7.5].groupby('Country')['Happiness Score'].mean()
```

```python
# Set the figure size
plt.figure(figsize=(8, 6))

# Create the bar plot
plt.bar(grouped.index, grouped.values)

# Customize the plot
plt.xlabel('Country')
plt.ylabel('Average Happiness Score')
plt.title('Average Happiness Score by Country')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
life = df[df['Health (Life Expectancy)'] >= 0.95].groupby('Country')['Health (Life Expectancy)'].mean()
```

```python
plt.figure(figsize = (8,6))

plt.barh(life.index, life.values)

plt.xlabel('Country')
plt.ylabel('Health (Life Expectancy)')
plt.title('Life Expectancy by Country')

plt.show()
```

```python
test_avg = df.groupby(['Region']).mean(numeric_only=True).sort_values('Happiness Score', ascending=False)
```

```python
ax1 = test_avg.plot( y=['Economy (GDP per Capita)','Family', 'Health (Life Expectancy)',
                           'Freedom','Trust (Government Corruption)','Generosity'],
                       figsize=(15,5), kind='bar', rot=45)
ax1.set_ylabel('Score')
ax1.set_title('Average Scores by Region')
plt.show()
```
